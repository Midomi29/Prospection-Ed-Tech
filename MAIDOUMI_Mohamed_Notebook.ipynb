#### Nétoyage des données Region et Country

```python
regions=['Arab World',
         'East Asia & Pacific',
         'East Asia & Pacific (excluding high income)',
         'East Asia & Pacific (developing only)',
         'East Asia & Pacific (all income levels)',
         'Euro area',
         'European Union',
         'Europe & Central Asia',
         'Europe & Central Asia (excluding high income)',
         'Europe & Central Asia (developing only)',
         'Europe & Central Asia (all income levels)',
         'European Union',
         'Heavily indebted poor countries (HIPC)',
         'High income',
         'Latin America & Caribbean',
         'Latin America & Caribbean (developing only)',
         'Latin America & Caribbean (excluding high income)',
         'Least developed countries: UN classification',
         'Low & middle income',
         'Low income',
         'Lower middle income',
         'Middle East & North Africa',
         'Middle East & North Africa (excluding high income)',
         'Middle East & North Africa (all income levels)',
         'Middle East & North Africa (developing only)',
         'Middle income',
         'North America',
         'OECD members',
         'South Asia',
         'Sub-Saharan Africa',
         'Sub-Saharan Africa (excluding high income)',
         'Upper middle income',
         'World',
        ]
```

```python
countries=data_[~data_['Country Name'].isin(regions)]['Country Name'].unique()
```

```python
import numpy as np

indicators= ['UIS.E.4',
              'UIS.E.3',
              'SE.TER.ENRL', #enrolement
              
              'UIS.XSPENDP.56.FDPUB.FNCAP', # Capital expenditure as % of total expenditure in tertiary public institutions (%)
              
              'SP.POP.1524.TO.UN', # pop 20-29
              
              'IT.NET.USER.P2', # internet users
             ]


## NaN per indicator 
for indicator in indicators:
    sns.set()
    labels = ['NAN', 'Data']
    sizes = [np.sum(data_[data_['Indicator Code'].isin([indicator])]['NaN']), 
    data_[data_['Indicator Code'].isin(["UIS.E.4"])][['Indicator Code']].count()-np.sum(data_[data_['Indicator Code'].isin(["UIS.E.4"])]['NaN'])]
    explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
    plt.title("{}".format(indicator))#data_[data_['Indicator Code'].isin([indicator])]['Indicator Name'].unique().tolist()))
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()

####### on recupere les data in our incators per country
countries=data_[~data_['Country Name'].isin(regions)]['Country Name'].unique()
data_country = data_[data_['Country Name'].isin(countries)]
data_country_ind=data_country[data_country['Indicator Code'].isin(indicators)]
data_country_ind.drop(['NaN'], axis=1)
```

```python
df1_ind= data_country_ind[data_country_ind['Indicator Code'].isin(['UIS.E.4'])][['2010-2015','Country Name']] 
df1_ind['score1']=(df1_ind['2010-2015']-df1_ind['2010-2015'].mean()).abs()/(df1_ind['2010-2015'].max()-df1_ind['2010-2015'].min())
df1_ind['score1'][df1_ind['score1'].isna()]=0

df2_ind= data_country_ind[data_country_ind['Indicator Code'].isin(['UIS.E.3'])][['2010-2015','Country Name']] 
df2_ind['score2']=(df2_ind['2010-2015']-df2_ind['2010-2015'].mean()).abs()/(df2_ind['2010-2015'].max()-df2_ind['2010-2015'].min())
df2_ind['score2'][df2_ind['score2'].isna()]=0


df3_ind= data_country_ind[data_country_ind['Indicator Code'].isin(['SE.TER.ENRL'])][['2010-2015','Country Name']] 
df3_ind['score3']=(df3_ind['2010-2015']-df3_ind['2010-2015'].mean()).abs()/(df3_ind['2010-2015'].max()-df3_ind['2010-2015'].min())
df3_ind['score3'][df3_ind['score3'].isna()]=0


df4_ind= data_country_ind[data_country_ind['Indicator Code'].isin(['UIS.XSPENDP.56.FDPUB.FNCAP'])][['2010-2015','Country Name']] 
df4_ind['score4']=(df4_ind['2010-2015']-df4_ind['2010-2015'].mean()).abs()/(df4_ind['2010-2015'].max()-df4_ind['2010-2015'].min())
df4_ind['score4'][df4_ind['score4'].isna()]=0


df5_ind= data_country_ind[data_country_ind['Indicator Code'].isin(['SP.POP.1524.TO.UN'])][['2010-2015','Country Name']] 
df5_ind['score5']=(df5_ind['2010-2015']-df5_ind['2010-2015'].mean()).abs()/(df5_ind['2010-2015'].max()-df5_ind['2010-2015'].min())
df5_ind['score5'][df5_ind['score5'].isna()]=0


df6_ind= data_country_ind[data_country_ind['Indicator Code'].isin(['IT.NET.USER.P2'])][['2010-2015','Country Name']] 
df6_ind['score6']=(df6_ind['2010-2015']-df6_ind['2010-2015'].mean()).abs()/(df6_ind['2010-2015'].max()-df6_ind['2010-2015'].min())
df6_ind['score6'][df6_ind['score6'].isna()]=0



df=df1_ind.merge(how='left', on='Country Name',right= df2_ind )
df=df.merge(how='left', on='Country Name',right= df3_ind )
df=df.merge(how='left', on='Country Name',right= df4_ind )
df=df.merge(how='left', on='Country Name',right= df5_ind )
df=df.merge(how='left', on='Country Name',right= df6_ind )

df['score_syn']=df['score1']+df['score2']+df['score3']+df['score4']+df['score5']+df['score6']

```

```python
df.sort_values(by="score_syn",ascending=False).iloc[:15]

plt.figure(figsize = (10,10))
sns.set()
ax = plt.gca()
sns.barplot(x = df.sort_values(by="score_syn",ascending=False).iloc[:15]['score_syn']
,y=df.sort_values(by="score_syn",ascending=False).iloc[:15]['Country Name'])
plt.title('', size=20)
#plt.xlabel('Nombre')
plt.ylabel('Pays')
plt.show()
```

```python
# indicateur lycéens + etudiants supérieur:
    
    
df['customers'] =df['score1'] +df['score2'] +df['score3']
print('liste des pays avec le plus de lycéens et étudiants')
df.sort_values(by='customers', ascending=False)[['Country Name', 'customers']].head(20)

plt.figure(figsize = (10,10))
sns.set()
ax = plt.gca()
sns.barplot(x = df.sort_values(by='customers', ascending=False)['customers'].head(20), y=df.sort_values(by='customers', ascending=False)['Country Name'].head(20))
plt.title('Nombre de lycéens et étudiants par pays', size=20)
#plt.xlabel('Nombre')
plt.ylabel('Pays')
plt.show()
```

```python
#on va multiplier l'estimatoeur des inscriptions lycée+supieur avec le score de pénétration d'internet pour avoir une estimation du nombre de clients potentiels:
df['potential_customers'] = df['customers'] * df['score6']
df.sort_values(by='potential_customers', ascending=False)[['Country Name', 'score6', 'customers', 'potential_customers']].head(20)
clients_potentiels = df.sort_values(by='potential_customers', na_position='first', ascending=True)[['Country Name', 'score6', 'customers', 'potential_customers']].tail(20)

plt.figure(figsize = (10,10))
sns.set()
plt.title('Nombre de clients potentiels par pays (20 premiers pays)', fontsize=20)
ax = plt.gca()
sns.barplot(x = clients_potentiels['potential_customers'], y=clients_potentiels['Country Name'])
plt.xlabel('Nombre de clients potentiels')
plt.ylabel('Pays')
plt.show()
```
